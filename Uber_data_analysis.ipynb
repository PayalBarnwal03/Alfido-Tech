# **UBER DATA ANALYSIS**

```python
#Importing Libraries
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
#Loading Dataset
df=pd.read_csv('/content/UberDataset.csv')
df=pd.DataFrame(df)
print(df)
```

```python
#Check the null values
df.isnull().sum()
```

```python
df["PURPOSE"]=df["PURPOSE"].fillna("MISSING")
df.dropna(inplace=True)
```

```python
df.dropna()
```

```python
df['START'].replace('Kar?chi','Karachi', inplace=True)
df['STOP'].replace('Kar?chi','Karachi', inplace=True)
```

```python
df['START'],df['STOP']
```

```python
df.nunique()
```

```python
df.dtypes
```

```python
df.drop_duplicates(subset=None, keep='first', inplace = True)
df.duplicated().sum()
```

```python
df.size
```

```python
df['START_DATE']=pd.to_datetime(df['START_DATE'])
df['START_DATE']
```

```python
df['month']=df['START_DATE'].dt.month
df['day']=df['START_DATE'].dt.day
```

```python
f=plt.figure()
f.set_figheight(7)
f.set_figwidth(20)
sns.countplot(x="CATEGORY", data=df)
```

```python
f=plt.figure()
f.set_figheight(7)
f.set_figwidth(20)
sns.histplot(x="PURPOSE", data=df)
```

```python
import seaborn as sns
sns.histplot(df)
```

```python
corr=df.corr()
sns.heatmap(corr,annot=True)
```

```python
sns.scatterplot(data=df, x="month",y="MILES", legend=False, sizes=(20, 200))

plt.show()
```

```python
f=plt.figure()
f.set_figheight(7)
f.set_figwidth(20)
sns.barplot(x='day',y='MILES' ,data=df )
```
